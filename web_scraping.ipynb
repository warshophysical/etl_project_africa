```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_African_countries_by_area"
```

```python
response=requests.get(url)
```

```python
soup=BeautifulSoup(response.text,'html.parser')
```

```python
print(soup.prettify())
```

```python
results=soup.tbody.find_all('a')
```

```python
results
```

```python
countryList=[]
for result in results:
    countryList.append(result.text)
```

```python
dfCountry=pd.DataFrame(data=countryList,columns=["Country"])
```

```python
print(dfCountry.info())
```

```python
dfCountry=dfCountry.drop([0, 1,2,3,4,5,18])
dfCountry
```

```python
dfCountry=dfCountry.reset_index(drop=True)
dfCountry
```

```python
dfCountry.loc[dfCountry['Country']=="Republic of the Congo","Country"] = "Congo"
dfCountry.loc[dfCountry['Country']=="Ivory Coast (Côte d'Ivoire)","Country"] = "Ivory Coast"
dfCountry.loc[dfCountry['Country']=="Eswatini (Swaziland)","Country"] = "Eswatini"
dfCountry.loc[dfCountry['Country']=="Gambia, The","Country"] = "Gambia"
dfCountry
```
